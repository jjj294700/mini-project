```python

plt.rcParams['font.family'] = 'Malgun Gothic'

file_map = {
    ''
    'kiJobByType.csv': 'occupation',
    'kiAlchbyType.csv': 'alcohol',
    'kiSmokeByType.csv': 'smoking_now',
    'kiPastSmokeByType.csv': 'smoking_past',
    'kiHighBloodByType.csv': 'hypertension',
    'kiDiabetesByType.csv': 'diabetes',
    'kiCaseByType.csv': 'past_cancer',
    'kiHeartByType.csv': 'heart_disease'
}

columns_to_drop = [
    'ncsNmvl', 'wholNcsDnmvl', 'ptCntNmvl',
    'wholPtCntDnmvl', 'statsMetaNo', 'centerNm', 'critYr'
]

# 결과 저장 디렉토리 만들기
os.makedirs('output_charts', exist_ok=True)
os.makedirs('output_csv', exist_ok=True)

for file_name, new_column_name in file_map.items():
    try:
        df = pd.read_csv(file_name)

        # 컬럼명 정리
        if 'ptAge' in df.columns:
            df.rename(columns={'ptAge': 'age'}, inplace=True)
        if 'ptSexCd' in df.columns:
            df.rename(columns={'ptSexCd': 'gender'}, inplace=True)
        if 'statsTrgtNm' in df.columns:
            df.rename(columns={'statsTrgtNm': new_column_name}, inplace=True)

        df.drop(columns=[col for col in columns_to_drop if col in df.columns], inplace=True)

        if 'age' in df.columns:
            df['age'] = df['age'].astype(str)
        if 'gender' in df.columns:
            df['gender'] = df['gender'].map({'M': '1', 'F': '0'})
        if 'occupation' in df.columns:
            df['occupation'] = df['occupation'].astype(str)
        if new_column_name in df.columns:
            df = df.dropna(subset=[new_column_name])  
            df[new_column_name] = df[new_column_name].astype(str)  

        # NaN 제거하고 value_counts
        series = df[new_column_name].dropna()
        value_counts = series.value_counts()

        # 빈도수 결과 저장
        freq_df = pd.DataFrame({new_column_name: value_counts.index, 'Frequency': value_counts.values})
        freq_df.to_csv(f'output_csv/freq_{new_column_name}.csv', index=False)

        # 막대그래프 저장
        plt.figure(figsize=(10, 6))
        bars = plt.bar(value_counts.index, value_counts.values, color='#69b3a2', edgecolor='black')

        # 그림자 느낌 추가
        for bar in bars:
            bar.set_alpha(0.85)
            bar.set_linewidth(1)

        plt.title(f"{new_column_name} 빈도수", fontsize=16)
        plt.xlabel(new_column_name, fontsize=12)
        plt.ylabel("Frequency", fontsize=12)
        plt.tight_layout()
        plt.savefig(f'output_charts/freq_{new_column_name}.png')
        plt.close()

        print(f"{file_name} → {new_column_name}: 완료")

    except FileNotFoundError:
        print(f"파일 없음: {file_name}")
    except Exception as e:
        print(f"에러 발생: {file_name} - {e}")

print("✅ 모든 처리 완료 (NaN 제거, CSV + 그래프 저장)")
```

```python
import pandas as pd

# CSV 파일 경로와 항목 맵핑
files = {
    'output_csv/freq_smoking_now.csv': 'smoking_now',
    'output_csv/freq_smoking_past.csv': 'smoking_past',
    'output_csv/freq_alcohol.csv': 'alcohol',
    'output_csv/freq_diabetes.csv': 'diabetes',
    'output_csv/freq_hypertension.csv': 'hypertension',
    'output_csv/freq_heart_disease.csv': 'heart_disease',
    'output_csv/freq_past_cancer.csv': 'past_cancer'
}

# 위험 점수 계산 함수
def calculate_risk(df, column_name):
    # Y와 N 응답 수 계산
    y_count = df[df[column_name] == 'Y']['Frequency'].sum()
    n_count = df[df[column_name] == 'N']['Frequency'].sum()

    # 총 응답 수 계산
    total_count = y_count + n_count
    
    if total_count == 0:
        return None  # 총 응답수가 0일 경우 처리
    
    # Y 비율 계산
    y_ratio = (y_count / total_count) * 100

    # 위험 점수 부여 기준
    if y_ratio > 40:
        risk_score = "+1점 (위험 있음)"
    elif 10 < y_ratio <= 40:
        risk_score = "+0.5점 (위험 보통)"
    else:
        risk_score = "+0.25점 (위험 낮음)"
    
    return y_count, n_count, f"{y_ratio:.2f}%", risk_score

# 결과를 저장할 리스트
results = []

# 파일 읽고 처리
for file_path, column_name in files.items():
    # CSV 파일에서 데이터 읽기
    df = pd.read_csv(file_path)
    
    # 위험 점수 계산
    result = calculate_risk(df, column_name)
    
    if result:
        y_count, n_count, y_ratio, risk_score = result
        results.append([column_name, y_count, n_count, y_ratio, risk_score])

# 결과를 DataFrame으로 변환
columns = ['항목', 'Y 응답 수', 'N 응답 수', 'Y 비율', '위험 점수 부여 기준']
df_risk = pd.DataFrame(results, columns=columns)

# 결과 출력
print(df_risk)

# 결과를 CSV로 저장
df_risk.to_csv('output_risk/output_data_with_risk_scores.csv', index=False)
```

```python
data = {
    'occupation': ['무직', '회사원', '기타', '자유업', '주부', 'nan', '전문직', '교사', '학생', '군인'],
    'Frequency': [221, 154, 141, 122, 117, 29, 23, 9, 2, 1]
}

df = pd.DataFrame(data)

# 'nan' 값 제외
df = df[df['occupation'] != 'nan']

# 전체 응답자 수 계산
total_count = df['Frequency'].sum()


def assign_risk_score(y_ratio):
    if y_ratio >= 40:
        return "+1점 (위험 있음)"
    elif 10 <= y_ratio < 40:
        return "+0.5점 (위험 보통)"
    else:
        return "+0.25점 (위험 낮음)"

# 결과를 저장할 리스트
risk_data = []

# occupation 항목에 대한 비율 계산 및 위험 점수 부여
for _, row in df.iterrows():
    occupation = row['occupation']
    frequency = row['Frequency']
    n_frequency = total_count - frequency  # N 응답 수
    y_ratio = (frequency / total_count) * 100  # 각 직업의 비율 계산
    risk_score = assign_risk_score(y_ratio)  # 위험 점수 부여
    risk_data.append([occupation, frequency, n_frequency, f"{y_ratio:.2f}%", risk_score])

# DataFrame으로 변환
columns = ['항목', '응답 수', 'N 응답 수', '비율', '위험 점수 부여 기준']
df_risk = pd.DataFrame(risk_data, columns=columns)

# 결과 출력
print(df_risk)

# CSV로 저장
df_risk.to_csv('output_risk/occupation_risk_scores.csv', index=False)
```
