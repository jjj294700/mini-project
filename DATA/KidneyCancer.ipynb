```python
for api in apis:  # save information as json file
    url = api['url']
    params = api['params']
    
    page_no = 1  # Start from the first page
    all_data = []  # To store all data
    
    while True:
        params['pageNo'] = str(page_no)  # Set the current page
        response = requests.get(url, params=params)
        
        if response.status_code == 200:
            json_data = response.json()
            
            if 'items' in json_data:  # Check if 'items' exist in the response
                data = json_data['items']
                all_data.extend(data)  # Add the current page's data to the list
                
                # If the number of records is less than 'numOfRows', stop pagination
                if len(data) < int(params['numOfRows']):
                    break
                page_no += 1  # Move to the next page
            else:
                print(f"No items found in the response for {url}")
                break
        else:
            print(f"Error: {response.status_code} for {url}")
            break
    
    # Save all collected data as a JSON file
    filename = url.split('/')[-1] + '.json'
    with open(filename, 'w', encoding='utf-8') as json_file:
        json.dump(all_data, json_file, ensure_ascii=False, indent=4)
    
    print(f"Data from {url} saved as JSON.")
```

```python
import pandas as pd

# 파일 이름과 바꿀 컬럼 이름 매핑
file_map = {
    'kiJobByType.csv': 'occupation',
    'kiAlchbyType.csv': 'alcohol',
    'kiSmokeByType.csv': 'smoking_now',
    'kiPastSmokeByType.csv': 'smoking_past',
    'kiHighBloodByType.csv': 'hypertension',
    'kiDiabetesByType.csv': 'diabetes',
    'kiCaseByType.csv': 'cancer',
    'kiHeartByType.csv': 'heart_disease'
}

# 삭제할 컬럼 목록
columns_to_drop = [
    'ncsNmvl', 'wholNcsDnmvl', 'ptCntNmvl',
    'wholPtCntDnmvl', 'statsMetaNo', 'centerNm', 'critYr'
]

# 각 파일에 대해 개별적으로 처리
for file_name, new_column_name in file_map.items():
    try:
        # CSV 파일 읽기
        df = pd.read_csv(file_name)

        # ptAge를 age로 변환, ptSexCd를 gender로 변환
        if 'ptAge' in df.columns:
            df.rename(columns={'ptAge': 'age'}, inplace=True)

        if 'ptSexCd' in df.columns:
            df.rename(columns={'ptSexCd': 'gender'}, inplace=True)

        # 컬럼 이름 변경 (statsTrgtNm 컬럼을 'new_column_name'으로 변경)
        if 'statsTrgtNm' in df.columns:
            df.rename(columns={'statsTrgtNm': new_column_name}, inplace=True)

        # 필요 없는 컬럼 삭제 (존재할 때만 삭제)
        df.drop(columns=[col for col in columns_to_drop if col in df.columns], inplace=True)

        # 숫자형 데이터를 문자열로 변환
        if 'age' in df.columns:
            df['age'] = df['age'].astype(str)  # age를 문자열로 변환

        if 'gender' in df.columns:
            # 'M'을 '1', 'F'를 '0'으로 변환
            df['gender'] = df['gender'].map({'M': '1', 'F': '0'})  # 'M'을 '1', 'F'를 '0'으로 변환

        # 'occupation'은 문자열로 변환
        if 'occupation' in df.columns:
            df['occupation'] = df['occupation'].astype(str)

        # statsTrgtNm을 문자열로 변환 (혹시 numeric 형식이라면 이를 처리)
        if new_column_name in df.columns:
            df[new_column_name] = df[new_column_name].astype(str)

        # 저장
        output_file = f'cleaned_{new_column_name}.csv'
        df.to_csv(output_file, index=False)
        print(f" {file_name} → '{output_file}' 저장 완료 (컬럼 이름 변경 + 불필요 컬럼 제거)")

    except FileNotFoundError:
        print(f" 파일 없음: {file_name}")
    except Exception as e:
        print(f" 에러 발생: {file_name} - {e}")

print("모든 파일에 대한 처리 완료")

```

```python
from functools import reduce
from scipy.stats import chi2_contingency

file_map = {
    'sorted_kiAlchbyType.csv': 'alcohol',
    'sorted_kiSmokeByType.csv': 'smoking_now',
    'sorted_kiPastSmokeByType.csv': 'smoking_past',
    'sorted_kiHighBloodByType.csv': 'hypertension',
    'sorted_kiDiabetesByType.csv': 'diabetes',
    'sorted_kiCaseByType.csv': 'cancer',
    'sorted_kiHeartByType.csv': 'heart_disease'
}

dfs = []

for file, col_name in file_map.items():
    df = pd.read_csv(file)
    
    # 맨 마지막 컬럼 이름 바꾸기
    df = df.rename(columns={df.columns[-1]: col_name})
    
    dfs.append(df)

# statsMetaNo, age, gender 기준 merge
merged_df = reduce(lambda left, right: pd.merge(left, right, on=['statsMetaNo', 'age', 'gender'], how='outer'), dfs)

# Y/N → 1/0 변환
for col in file_map.values():
    merged_df[col] = merged_df[col].map({'Y': 1, 'N': 0})

print(merged_df.head())
```

```python
# gender 컬럼 숫자로 변환
merged_df['gender'] = merged_df['gender'].map({'M': 1, 'F': 0})

# NaN -> 0 처리
merged_df = merged_df.fillna(0)

# 상관관계 분석
correlation = merged_df.corr()

print(correlation)

```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# 숫자형 컬럼만 평균값으로 NaN 채우기
merged_df = merged_df.fillna(merged_df.mean(numeric_only=True))

# 상관계수 저장
correlation.to_csv('correlation_result_meanfill.csv', encoding='utf-8-sig')

print("===== 상관계수 분석 결과 =====")
print(correlation)

# 히트맵 그리기
plt.figure(figsize=(10, 8))
sns.heatmap(correlation, annot=True, fmt=".2f", cmap='coolwarm')
plt.title('Correlation Heatmap (NaN -> Mean Fill)')
plt.tight_layout()
plt.savefig('correlation_heatmap_meanfill.png')  # 이미지 저장
plt.show()
```
